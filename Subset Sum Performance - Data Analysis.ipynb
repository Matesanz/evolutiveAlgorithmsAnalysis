```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# Modificamos el tamaño predeterminado de los gráficos

sns.set(rc={'figure.figsize':(32,32)})
rcParams['figure.figsize'] = 11.7,8.27
```

```python
#Importamos el dataset obtenido de de los resultados del algoritmo

df = pd.read_csv("./subsetSumPerformance.csv", header=None)
df
```

```python
# Redimensión del Dataset obtenido

df_array = df.values.astype(int)
df_array = df_array.reshape((9, 8, 50))
```

```python
# Dividir el número de iteraciones por el tamaño de su población

df_weihted_array = df_array / np.array([2,4,8,16,32,64,128,256,512])[:,np.newaxis,np.newaxis]
```

```python
# Creamos los directorios donde irán los archivos resultantes de este ejercicio

if not os.path.exists("excel/"):
    os.mkdir("excel/")
if not os.path.exists("graphs/subsetSum/"):
    os.mkdir("graphs/subsetSum/")
```

```python
# Mediana de las distintas rondas de ejecución

row_names= ['Población 2', 'Población 4', 'Población 8', 'Población 16', 'Población 32', 'Población 64', 'Población 128', 'Población 256', 'Población 512']
column_names = ['Gn: 1 | Gl: 128', 'Gn: 2 | Gl: 64', 'Gn: 4 | Gl: 32', 'Gn: 8 | Gl: 16', 'Gn: 16 | Gl: 8', 'Gn: 32  | Gl: 4', 'Gn: 64 | Gl: 2', 'Gn: 128 | Gl: 1']


np_median = np.median(df_weihted_array, axis=2)
df_median = pd.DataFrame(np_median, index=row_names, columns=column_names)
df_median.to_excel("excel/median_Subset_Sum.xlsx")
df_median
```

```python
# Desviavión Estándar de las distintas rondas de ejecución

np_std = np.std(df_weihted_array, axis=2)
df_std = pd.DataFrame(np_std, index=row_names, columns=column_names )
df_std.to_excel("excel/std_Deviation_Subset_Sum.xlsx")
df_std
```

```python
# Nuevo Dataframe categorizado por columnas

flat = df_array.flatten()
new = np.ndarray((3600,5), dtype=int)
index=-1
for num_genes in range(0, 8):
    for num_poblacion in range(1, 10):
        for value in range(1, 51):
            index += 1
            new[index][0] = 2**num_poblacion
            new[index][1] = 2**num_genes
            new[index][2] = 2**(7 - num_genes)
            new[index][3] = flat[index]
            new[index][4] = flat[index]/(2**(num_poblacion))

column_names = ["Población", "Número Genes", "Longitud Genes", "Evaluaciones", "Iteraciones"]
new_df = pd.DataFrame(new, columns=column_names)

new_df.head()
```

```python
# Resultados Número de Iteraciones según Población

fig = sns.catplot(x="Población", y="Iteraciones", kind="bar", estimator=np.median ,data=new_df)
fig.set(yscale="log")
plt.title('Número de Iteraciones según Población')
fig.savefig('graphs/subsetSum/Iterations_vs_Individuos.png')
```

```python
# Resultados Número de Iteraciones según Longitud de Genes

fig = sns.catplot(x="Longitud Genes", y="Iteraciones", kind="bar", estimator=np.median ,data=new_df)
fig.set(yscale="log")
plt.title('Número de Iteraciones según Longitud de Genes')
fig.savefig('graphs/subsetSum/Iterations_vs_GL.png)
```

```python
# Resultados 'Número de Iteraciones según Número de Genes'

fig = sns.catplot(x="Número Genes", y="Iteraciones", kind="bar", estimator=np.median ,data=new_df)
fig.set(yscale="log")
plt.title('Número de Iteraciones según Número de Genes')
fig.savefig('graphs/subsetSum/Iterations_vs_GN.png')
```

```python
# Resultados 'Iteraciones según Población y Número de Genes'

fig = sns.catplot(x="Población", y="Iteraciones", hue="Número Genes", kind="bar", estimator=np.median ,data=new_df)
fig.fig.set_size_inches(20,7)
fig.set(yscale="log")
plt.title('Iteraciones según Población y Número de Genes')
fig.savefig('graphs/subsetSum/Iterations_vs_Individuos_vs_GNpng')
```

```python
# Número de iteraciones según número de genes por población. Escala no logarítmica

for i in range(1,10):    
    new_df_pob =  new_df[new_df['Población']==(2**i)]
    fig = sns.catplot(x="Número Genes", y="Iteraciones", kind="bar", legend_out=True, estimator=np.median, data=new_df_pob)
    # fig.set(yscale="log");
    plt.title('Población ' + str(2**i) + ' Individuos')
    fig.savefig('graphs/subsetSum/Poblacion_' + str(2**i) + '_Individuos.png')
```
