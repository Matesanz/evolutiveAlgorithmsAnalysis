```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
#Tamaño de los gráficos
sns.set(rc={'figure.figsize':(32,32)})
rcParams['figure.figsize'] = 11.7,8.27
```

```python
#Importamos el dataset obtenido de de los resultados del algoritmo

df = pd.read_csv("./oneMaxPerformance.csv", header=None)
df = df.dropna()
df.head()
```

```python
# Redimensión del Dataset obtenido

df_array = df.values.astype(int)
df_array = df_array.reshape((9, 9, 50))
```

```python
# Mediana de las distintas rondas de ejecución

row_names= ['Población 2', 'Población 4', 'Población 8', 'Población 16', 'Población 32', 'Población 64', 'Población 128', 'Población 256', 'Población 512']
column_names = ['Long. Gen: 2', 'Long. Gen: 4', 'Long. Gen: 8', 'Long. Gen: 16', 'Long. Gen: 32', 'Long. Gen 64', 'Long. Gen 128', 'Long. Gen 256', 'Long. Gen 512']
np_median = np.median(df_array, axis=2)
df_median = pd.DataFrame(np_median, index=row_names, columns=column_names )
df_median
```

```python
# Mediana de las distintas rondas de ejecución

row_names= ['Población 2', 'Población 4', 'Población 8', 'Población 16', 'Población 32', 'Población 64', 'Población 128', 'Población 256', 'Población 512']
column_names = ['Long. Gen: 2', 'Long. Gen: 4', 'Long. Gen: 8', 'Long. Gen: 16', 'Long. Gen: 32', 'Long. Gen 64', 'Long. Gen 128', 'Long. Gen 256', 'Long. Gen 512']
np_std = np.std(df_array, axis=2)
df_std = pd.DataFrame(np_std, index=row_names, columns=column_names )
df_std
```

```python
# Nuevo Dataframe categorizado por columnas

flat = df_array.flatten()
new = np.ndarray((4050,4), dtype=int)
index=-1
for num_genes in range(1, 10):
    for num_poblacion in range(1, 10):
        for value in range(1, 51):
            index += 1
            new[index][1] = 2**num_poblacion
            new[index][0] = 2**num_genes
            new[index][2] = flat[index]
            new[index][3] = flat[index]/(2**(num_genes))

column_names = ["Individuos", "Genes", "Evaluaciones", "Iteraciones"]
new_df = pd.DataFrame(new, columns=column_names)
new_df.head()
```

```python
# Resultados 
fig = sns.catplot(x="Individuos", y="Iteraciones", hue="Genes", kind="bar", estimator=np.median ,data=new_df)
fig.fig.set_size_inches(15,15)
fig.set(yscale="log")
```

```python
# Escala no logarítmica
for i in range(1,10):    
    new_df_pob =  new_df[new_df['Individuos']==(2**i)]
    fig = sns.catplot(x="Genes", y="Iteraciones", kind="bar", legend_out=True, estimator=np.median, data=new_df_pob)
    fig.set(yscale="log");
    plt.title('Población ' + str(2**i) + ' Individuos')
    fig.savefig('Poblacion ' + str(2**i) + ' Individuos.png')
```
